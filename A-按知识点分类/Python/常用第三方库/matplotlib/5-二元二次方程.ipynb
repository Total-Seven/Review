```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-3,3,100)  # 区间
y = 9 - x**2 

plt.xlim(-3,3)                 # X轴范围
plt.xticks(list(range(-3,4)))  # X轴刻度
plt.ylim(0.0,9.0)              # Y轴范围

plt.plot(x,y)

ax = plt.gca()          # 当前轴 获取
ax.set_aspect('equal')  # 横纵比 相等
ax.grid(True)           # 网格 显示

plt.show()
```
