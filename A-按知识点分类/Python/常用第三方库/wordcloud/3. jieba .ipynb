```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
import jieba
```

```python
f = open("word.txt").read()
```

#### wordcloud无法直接生成正确的中文词云

```python
cuted_f = "".join( jieba.lcut(f) )
```

```python
wordcloud = WordCloud(font_path = "C:/Windows/Fonts/simfang.ttf",background_color="white",width=1000,height=880).generate(cuted_f)
```

```python
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```
