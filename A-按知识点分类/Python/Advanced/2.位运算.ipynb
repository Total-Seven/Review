# `位`运算

> 世界上只有10种人，懂二进制的和不懂二进制的。

***

> 与运算（AND）： 将两个二进制数的对应位进行逻辑与操作，只有当两个位都是1时，结果才是1。用符号 "&" 表示。

|1010 (十进制 10)|
|-------|
| 1100 (十进制 12)|
|与|
|1000 (十进制 8)|

> 或运算（OR）： 将两个二进制数的对应位进行逻辑或操作，只要两个位中有一个是1，结果就是1。用符号 "|" 表示。

|1010 (十进制 10)|
|------------|
| 1100 (十进制 12)|
|或|
|1110 (十进制 14)|

> 异或运算（XOR）： 将两个二进制数的对应位进行逻辑异或操作，如果两个位不相同，则结果是1；如果相同，则结果是0。用符号 "^" 表示。
 
|1010 (十进制 10)|
|---------|
|1100 (十进制 12)|
|异或|
|0110 (十进制 6)|

> 取反运算（NOT）： 对一个二进制数的每一位取反，即0变成1，1变成0。用符号 "~" 表示。

|1010 (十进制 10)|
|---------|
|取反|
|0101 (十进制 -11，因为取反后加一)|
